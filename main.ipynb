```python
import os
import re

import pandas as pd

def preprocess_df_name(df_temp):
  count_l = []
  day_l = []
  for i, item in df_temp.iterrows():
    count_l.append(item['カテゴリ: すべてのカテゴリ'])
    day_l.append(i)
  name = re.findall(r"(.+?)[:.+]", count_l[0])[0]
  df_return = pd.DataFrame({'datetime': day_l[1:], name: count_l[1:]})
  return df_return
```

```python
cd_list = os.listdir('./data')
df_main = pd.read_csv('data/'+cd_list[0])
df_main = preprocess_df_name(df_main)
for data_name in cd_list[1:]:
    path = 'data/' + data_name
    df_temp = pd.read_csv(path)
    df_temp = preprocess_df_name(df_temp)
    append_name = df_temp.columns.values[-1]
    df_main[append_name] = df_temp[append_name]
```
